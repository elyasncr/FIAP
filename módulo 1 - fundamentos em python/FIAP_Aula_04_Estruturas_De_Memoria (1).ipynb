# Estruturas de Memória

## Introdução

### Memória


A memória é uma parte essencial na programação de computadores, sendo responsável por armazenar e recuperar dados durante a execução de programas. A organização eficiente da memória é crucial para otimizar o desempenho do sistema.

### Memória na Programação



Na programação, a memória refere-se aos locais onde os dados são armazenados temporariamente durante a execução de um programa. Isso inclui variáveis, estruturas de dados e outras informações necessárias para o funcionamento do programa.

- A memória em um sistema de computador.

    Em um sistema de computador, a memória pode ser dividida em diferentes tipos, como memória RAM (Random Access Memory) e memória de armazenamento permanente. A RAM é usada para armazenar dados temporários enquanto o computador está ligado, enquanto a memória de armazenamento permanente, como um disco rígido, é usada para armazenar dados de forma duradoura.


- A importância da organização eficiente da memória.

    A forma como a memória é organizada afeta diretamente o desempenho do sistema. Uma organização eficiente da memória permite um acesso rápido e otimizado aos dados, reduzindo o tempo de execução dos programas.

## Filas

### Definição
- Uma fila é uma estrutura de dados linear que segue a ordem FIFO (First In, First Out).
- Isso significa que o primeiro elemento a ser inserido na fila será o primeiro a ser removido.

### Operações Básicas
- Enfileirar (enqueue): Adicionar um elemento ao final da fila.
- Desenfileirar (dequeue): Remover o elemento do início da fila.



### Implementação em Python

```python
# Definição da classe Fila - FIFO - o primeiro elemento a ser inserido na fila será o primeiro a ser removido.
class Fila:
    # Método construtor (__init__) que inicializa a fila com uma lista vazia
    def __init__(self):
        self.items = []

    # Método para adicionar um item ao final da fila
    def enfileirar(self, item):
        self.items.append(item)

    # Método para remover e retornar o primeiro item da fila
    def desenfileirar(self):
        # Verifica se a fila não está vazia antes de tentar desenfileirar
        if not self.esta_vazia():
            # Remove e retorna o primeiro item da fila (pop(0) remove o primeiro elemento)
            return self.items.pop(0)

    # Método para verificar se a fila está vazia
    def esta_vazia(self):
        # Retorna True se o comprimento da lista (número de elementos) for igual a zero
        return len(self.items) == 0
```

```python
# Criando uma instância da classe Fila
fila = Fila()

# Verificando se a fila está vazia (deve retornar True)
print(fila.esta_vazia())

# Enfileirando alguns elementos na fila
fila.enfileirar(10)
fila.enfileirar(20)
fila.enfileirar(30)

# Imprimindo a fila
print(fila.items)
```

```python
# Verificando se a fila está vazia agora (deve retornar False)
print(fila.esta_vazia())
```

```python
# Desenfileirando um elemento (deve retornar 10, pois foi o primeiro a ser adicionado)
print(fila.desenfileirar())
```

```python
# Verificando o estado atual da fila após desenfileirar (deve conter [20, 30])
print(fila.items)
```

## Pilhas

### Definição
- Uma pilha é uma estrutura de dados linear que segue a ordem LIFO (Last In, First Out).
- Isso significa que o último elemento adicionado à pilha é o primeiro a ser removido.

### Operações Básicas

- Empilhar (push): Adicionar um elemento ao topo da pilha.
- Desempilhar (pop): Remover o elemento do topo da pilha.

###Implementação em Python

```python
# Definição da classe Pilha - LIFO - o último elemento adicionado à pilha é o primeiro a ser removido.
class Pilha:
    # Método construtor (__init__) que inicializa a pilha com uma lista vazia
    def __init__(self):
        self.items = []

    # Método para adicionar um item ao topo da pilha
    def empilhar(self, item):
        self.items.append(item)

    # Método para remover e retornar o item do topo da pilha
    def desempilhar(self):
        # Verifica se a pilha não está vazia antes de tentar desempilhar
        if not self.esta_vazia():
            # Remove e retorna o último item da lista (pop() sem argumentos remove o último elemento)
            return self.items.pop()

    # Método para verificar se a pilha está vazia
    def esta_vazia(self):
        # Retorna True se o comprimento da lista (número de elementos) for igual a zero
        return len(self.items) == 0
```

```python
# Criando uma instância da classe Pilha
pilha = Pilha()

# Verificando se a pilha está vazia (deve retornar True)
print(pilha.esta_vazia())

# Empilhando alguns elementos
pilha.empilhar(10)
pilha.empilhar(20)
pilha.empilhar(30)

# Imprimindo a pilha
print(pilha.items)
```

```python
# Verificando se a pilha está vazia agora (deve retornar False)
print(pilha.esta_vazia())
```

```python
# Desempilhando um elemento (deve retornar 30, pois foi o último a ser empilhado)
print(pilha.desempilhar())
```

```python
# Verificando o estado atual da pilha após desempilhar (deve conter [10, 20])
print(pilha.items)
```

## Matrizes

###Definição
- Uma matriz é uma estrutura de dados bidimensional composta por linhas e colunas.
- Cada elemento é identificado por seus índices de linha e coluna.


###Operações Básicas
- Adição e Multiplicação de Matrizes.

image.png

Em termos de alocação de memória, a matriz é armazenada de maneira contígua na memória. Cada elemento (número) na matriz é armazenado em uma posição específica da memória, e a estrutura de listas aninhadas mantém a organização dos elementos em linhas e colunas.

###Implementação em Python

```python
class Matriz:
    def __init__(self, linhas, colunas):
        # Inicializa a matriz com todas as entradas zeradas
        self.matriz = [[0] * colunas for _ in range(linhas)]
        self.linhas = linhas
        self.colunas = colunas

    def definir_elemento(self, linha, coluna, valor):
        # Define o valor de um elemento na matriz
        if 0 <= linha < self.linhas and 0 <= coluna < self.colunas:
            self.matriz[linha][coluna] = valor
        else:
            print("Índices fora do intervalo da matriz.")

    def obter_elemento(self, linha, coluna):
        # Obtém o valor de um elemento na matriz
        if 0 <= linha < self.linhas and 0 <= coluna < self.colunas:
            return self.matriz[linha][coluna]
        else:
            print("Índices fora do intervalo da matriz.")
            return None

    def imprimir_matriz(self):
        # Imprime a matriz
        for linha in self.matriz:
            print(linha)

# Criando uma instância da classe Matriz
matriz = Matriz(3, 3)

# Imprimindo a matriz
matriz.imprimir_matriz()
```

```python
# Definindo alguns elementos na matriz
matriz.definir_elemento(0, 0, 1)
matriz.definir_elemento(0, 1, 2)
matriz.definir_elemento(1, 1, 3)
matriz.definir_elemento(2, 2, 4)

# Imprimindo a matriz
matriz.imprimir_matriz()
```

```python
# Obtendo um elemento da matriz
elemento = matriz.obter_elemento(0, 0)
print(f"Elemento na posição (0, 0): {elemento}")
```

1. Construir uma classe em Python para representar uma fila (Queue) e realizar operações básicas como enfileirar, desenfileirar e verificar se a fila está vazia.




```python
class Fila:
    """
    A fila é utilizada para representar uma lista de tarefas a serem executadas.
    A adição de tarefas é feita usando enfileirar, e a conclusão de tarefas é feita usando desenfileirar
    """
    def __init__(self):
        self.items = []

    def enfileirar(self, item):
        # Adiciona o item ao final da fila
        self.items.append(item)

    def desenfileirar(self):
        # Remove e retorna o primeiro item da fila
        if not self.esta_vazia():
            return self.items.pop(0)

    def esta_vazia(self):
        # Verifica se a fila está vazia
        return len(self.items) == 0

# Criando uma instância da classe Fila
fila_tarefas = Fila()

# Adicionando tarefas à fila
fila_tarefas.enfileirar("Ler e-mails")
fila_tarefas.enfileirar("Escrever relatório")
fila_tarefas.enfileirar("Preparar apresentação")

# Imprimindo a fila original
print("Fila de tarefas original:", fila_tarefas.items)
```

```python
# Executando a primeira tarefa da fila
tarefa_concluida = fila_tarefas.desenfileirar()
print("Tarefa concluída:", tarefa_concluida)
```

```python
# Imprimindo a fila após a conclusão da primeira tarefa
print("Fila de tarefas após conclusão:", fila_tarefas.items)
```

2. Criar uma classe em Python para representar uma pilha (Stack) usando Programação Orientada a Objetos (POO), e implementar métodos para operações básicas.

```python
class Pilha:
    def __init__(self):
        # Inicializa a pilha com uma lista vazia
        self.items = []

    def empilhar(self, item):
        # Adiciona um elemento ao topo da pilha
        self.items.append(item)

    def desempilhar(self):
        # Remove e retorna o elemento do topo da pilha, se a pilha não estiver vazia
        if not self.esta_vazia():
            return self.items.pop()

    def esta_vazia(self):
        # Verifica se a pilha está vazia
        return len(self.items) == 0

# Criando uma instância da classe Pilha
pilha_exemplo = Pilha()

# Imprimindo a pilha
print("Pilha atual:", pilha_exemplo.items)
```

```python
# Removendo e imprimindo o elemento do topo
elemento_removido = pilha_exemplo.desempilhar()
print("Elemento removido do topo:", elemento_removido)
```

Desenvolva uma função para multiplicar duas matrizes.

```python
def multiplicar_matrizes(matriz1, matriz2):
    # Verifica se o número de colunas da primeira matriz é igual ao número de linhas da segunda matriz
    if matriz1.colunas != matriz2.linhas:
        raise ValueError("Número de colunas da primeira matriz deve ser igual ao número de linhas da segunda matriz.")

    # Cria uma matriz para armazenar o resultado da multiplicação
    resultado = Matriz(matriz1.linhas, matriz2.colunas)

    # Realiza a multiplicação das matrizes
    for i in range(matriz1.linhas):
        for j in range(matriz2.colunas):
            for k in range(matriz1.colunas):
                # Acumula o resultado da multiplicação na posição (i, j) da matriz resultado
                resultado.matriz[i][j] += matriz1.matriz[i][k] * matriz2.matriz[k][j]

    return resultado
```

```python
# Exemplo de uso
matriz1_exemplo = Matriz(2, 3)
matriz1_exemplo.matriz = [[1, 2, 3], [4, 5, 6]]
print(matriz1_exemplo.matriz)
```

```python
matriz2_exemplo = Matriz(3, 2)
matriz2_exemplo.matriz = [[7, 8], [9, 10], [11, 12]]
print(matriz2_exemplo.matriz)
```

- A matriz matriz1_exemplo tem dimensões 2x3 (2 linhas e 3 colunas).


- A matriz matriz2_exemplo tem dimensões 3x2 (3 linhas e 2 colunas).

Para calcular a multiplicação, o resultado terá dimensões correspondentes ao número de linhas da primeira matriz e ao número de colunas da segunda matriz, ou seja, uma matriz 2x2.

### Definindo as matrizes de exemplo

    matriz2_exemplo = [[7, 8], [9, 10], [11, 12]]


### Cálculo dos elementos da matriz resultante
### Para cada posição (i, j) na matriz resultante
    matriz_resultante[0][0] = 1*7 + 2*9 + 3*11
    matriz_resultante[0][1] = 1*8 + 2*10 + 3*12
    matriz_resultante[1][0] = 4*7 + 5*9 + 6*11
    matriz_resultante[1][1] = 4*8 + 5*10 + 6*12

### A matriz resultante após a multiplicação
    matriz_resultante = [[58, 64], [139, 154]]


```python
# Chama a função para multiplicar as matrizes
resultado_multiplicacao = multiplicar_matrizes(matriz1_exemplo, matriz2_exemplo)
```
